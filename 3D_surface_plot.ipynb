```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata
from matplotlib.ticker import MaxNLocator, ScalarFormatter


###spend ~ heat 3D surface plot for all places
# Replace 'path/to/your/file.csv' with the actual path to your CSV file
file_path = "E:/POI_gridmet_daily_weather_data//place_spend_merge_specific.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

plt.rcParams['axes.grid'] = True  # Turn on grid lines
plt.rcParams['grid.linewidth'] = 0.2  # Adjust grid line thickness (reduce it for thinner lines)


fig = plt.figure(figsize=(12, 10))

ax = fig.add_subplot(111, projection='3d')

# Create a grid
x_vals = np.linspace(df['ln_income'].min(), df['ln_income'].max(), 50)
y_vals = np.linspace(df['mean_heat'].min(), df['mean_heat'].max(), 50)

X, Y = np.meshgrid(x_vals, y_vals)

# Interpolate 'coeffs' onto the grid
Z = griddata(( df['ln_income'], df['mean_heat']), df['coeffs'], (X, Y), method='linear')



surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='k', linewidth=0.2)

# Add colorbar with custom color
#colorbar = fig.colorbar(surf, ax=ax, shrink=0.4, aspect=10, anchor=(0.5, 0.5))  # Adjust the anchor position

colorbar = fig.colorbar(surf, ax=ax, shrink=0.4, aspect=10, anchor=(0.5, 0.5), 
                        cmap='coolwarm', 
                        ticks=[0, -1, -2, -3, -4, -5])  # Change 'Custom Colorbar Label' and color as needed


# Set labels
ax.set_xlabel('ln(Income)', labelpad=20)
ax.set_ylabel('Extreme heat', labelpad=20)
ax.set_zlabel('Response sensibility', labelpad=20)

# Set background color to white
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))

# Customize grid lines (reduce the number of grid lines)

# Adjust x-axis tick label properties (font size and font weight)
ax.tick_params(axis='x', labelsize=8)
for tick in ax.get_xticklabels():
    tick.set_fontweight('normal')  # Adjust font weight here

ax.tick_params(axis='y', labelsize=8)
for tick in ax.get_xticklabels():
    tick.set_fontweight('normal')  # Adjust font weight here

ax.tick_params(axis='z', labelsize=8)
for tick in ax.get_xticklabels():
    tick.set_fontweight('normal')  # Adjust font weight here
    
ax.xaxis.set_major_locator(MaxNLocator(nbins=5))  # Change the number of bins as needed
ax.xaxis.set_major_formatter(ScalarFormatter(useOffset=False, useMathText=False))

ax.yaxis.set_major_locator(MaxNLocator(nbins=5))  # Change the number of bins as needed
ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=False, useMathText=False))

ax.zaxis.set_major_locator(MaxNLocator(nbins=5))  # Change the number of bins as needed
ax.zaxis.set_major_formatter(ScalarFormatter(useOffset=False, useMathText=False))

# Customize Matplotlib rcParams for thinner grid lines and smaller axis labels
#plt.rcParams['xtick.labelsize'] = 12  # Adjust X-axis label font size
#plt.rcParams['ytick.labelsize'] = 12  # Adjust Y-axis label font size
#plt.rcParams['ztick.labelsize'] = 13  # Adjust Z-axis label font size

#plt.savefig('C:/Users/zhaiw/Dropbox (Personal)/working paper/Heat_spend/figures/figure4/visit_surface_plot.svg', format='svg')

plt.show()
```
